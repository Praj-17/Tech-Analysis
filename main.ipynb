```python
import yfinance as yf

# Retrieve list of exchanges
exchanges = yf.Ticker('^BSESN').info['exchange']

# Filter for Indian exchanges
indian_exchanges = [exchange for exchange in exchanges if 'India' in exchange]

# Retrieve list of symbols for each exchange
indian_symbols = []
for exchange in indian_exchanges:
    ticker_list = yf.Tickers().tickers_on_exchange(exchange)
    indian_symbols.extend([ticker.info['symbol'] for ticker in ticker_list])

# Retrieve stock names
stock_names = {symbol: yf.Ticker(symbol).info['longName'] for symbol in indian_symbols}

# Print the results
print("Indian Stock Exchanges:", indian_exchanges)
print("Stock Symbols:", indian_symbols)
print("Stock Names:", stock_names)
```
